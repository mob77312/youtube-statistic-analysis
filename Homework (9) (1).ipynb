Jelaskan Fitur mana yang sebaiknya digunakan dari hasil EDA?

Fitur yang digunakan adalah `category_id`, `publish_time`, `comments_disabled`, `ratings_disabled`, `video_error_or_removed`, `likes`, `dislikes`, `No_tags`, `desc_len`, `len_title`. <br>
`category_id` : digunakan untuk mengetahui potensi category apa yang dapat mempengaruhi jenis view, karena berpengaruh terhadap jenis konten yang dibuat. <br>
`publish_time` : digunakan untuk mengetahui waktu mana yang memiliki potensi untuk mendapatkan view yang lebih banyak<br>
`comments_disabled` : digunakan untuk mengetahui seberapa pengaruh komentar yang di tutup terhadap view yang ada<br>
`ratings_disabled` : digunakan untuk mengetahui seberapa pengaruh penilaian yang di tutup terhadap view yang ada<br>
`video_error_or_removed` : digunakan untuk mengetahui apakah ada tidaknya pengaruh yang di hasilkan dari video error atau di hapus terhadap view<br>
`likes` : digunakan untuk mengetahui apakah ada hubungan terhadap jumlah like dengan jumlah view<br>
`dislikes` : digunakan untuk mengetahui apakah ada hubungan terhadap jumlah dislike dengan jumlah view<br>
`No_tags` : digunakan untuk mempengaruhi apakah ada hubungannya terhadap jumlah tag yang digunakan pada setiap video dengan views.<br>
`desc_len` : apakah adanya pengaruh terhadap jumlah huruf di deskripsi terhadap jumlah views.<br>
`len_title` : apakah adanya pengaruh terhadap jumlah huruf di judul terhadap jumlah views.<br>

# Model

```python
x = df_clean.drop(columns=['views', 'trending_date', 'publish_date', 'publish_time']) # menggunakan semua feature kecuali target
y = df_clean['views'] 
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=1/3, random_state=42)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def eval_regression(model, pred, xtrain, ytrain, xtest, ytest):
    print("MAE: %.2f" % mean_absolute_error(ytest, pred)) # The MAE
    print("RMSE: %.2f" % mean_squared_error(ytest, pred, squared=False)) # The RMSE
    print('R2 score: %.2f' % r2_score(ytest, pred)) # Explained variance score: 1 is perfect prediction
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(xtrain, ytrain)
y_pred = LR.predict(xtest)
eval_regression(LR, y_pred, xtrain, ytrain, xtest, ytest)
```

LASSO

```python
from sklearn.linear_model import Lasso
lasso_model = Lasso()
lasso_model.fit(xtrain, ytrain)

pred = lasso_model.predict(xtest)
eval_regression(lasso_model, pred, xtrain, ytrain, xtest, ytest)
```
